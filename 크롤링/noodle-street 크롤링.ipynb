```python
driver = webdriver.Chrome()
url = "https://www.diningcode.com/list.dc?query=%EA%B0%95%EB%82%A8%EC%97%AD%20%EB%A9%B4%EC%9A%94%EB%A6%AC"
driver.get(url)

# 로딩 대기 시간을 설정 (적절한 대기 시간을 설정해야 합니다)
driver.implicitly_wait(3)

data = driver.find_element(By.CSS_SELECTOR, "button.sc-ldgPui")
data.click()
driver.implicitly_wait(3)

for i in range(6):
    data.send_keys(Keys.END)
    time.sleep(1.5)

# 'a.sc-ilxdoh' 요소 찾아서 리스트에 담기
restaurant_links = driver.find_elements(By.CSS_SELECTOR, 'a.sc-ilxdoh')

# 결과 출력
for link in restaurant_links:
    restaurant_url = link.get_attribute('href')
   
    # 각 가게의 상세 페이지로 이동
    driver.get(restaurant_url)
    
    # 영업시간 더보기 누르기 
    plus = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div.busi-hours > p.r > a")))
    plus.click()

    # 메뉴 정보 더보기 누르기
    plus2 = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div.menu-info > p.r > a")))
    plus2.click()

    # 이미지 URL 찾기
    img = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "li.btn-gallery-open > img")))
    print("이미지 URL:", img.get_attribute("src"))

    # 가게 이름 찾기
    name = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "h1.tit")))
    print("가게 이름:", name.text)

    # 평점 찾기
    rating = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "span.point")))
    print("평점:", rating.text)

    # 주소 찾기
    addr1 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "li.locat > a"))).text
    addr2 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "li.locat > span"))).text
    addr = addr1 + addr2
    print("주소:", addr)

    # 가게 전화번호 찾기
    tel = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "li.tel")))
    print("전화번호:", tel.text)

    # 주차 가능 여부
    parking = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "li.char"))).text
    print("주차 가능 여부:", parking)

    # 영업시간 찾기
    opening1 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.busi-hours-today"))).text
    opening2 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.busi-hours"))).text
    opening = opening1 + opening2
    print("영업시간:", opening)

    # 메뉴 찾기
    menu = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.menu-info")))
    print("메뉴:", menu.text)

    # 대기 시간을 사용하여 원하는 요소가 나타날 때까지 기다림
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.menu-info")))
    
    # 3초 대기 (추가적인 대기 시간이 필요한 경우 조절)
    time.sleep(3)
```
